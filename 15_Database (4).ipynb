```python
import csv
import mysql.connector
import json
from decimal import Decimal

# Step 1: Read Data from CSV
csv_file_path = 'telecom_data.csv'  # Update this path as needed

# Reading the CSV data using the csv module
with open(csv_file_path, mode='r') as file:
    reader = csv.reader(file)
    header = next(reader)  # Skip the header row
    data = [row for row in reader]

# Step 2: Insert Data into MySQL
# Database connection details
db_config = {
    'user': 'sql3726984',
    'password': '47U1YDn2w4',
    'host': 'sql3.freesqldatabase.com',
    'database': 'sql3726984'
}

# Connect to the database
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()

# Create table if not exists
create_table_query = """
CREATE TABLE IF NOT EXISTS telecom_data (
    customer_id INT PRIMARY KEY,
    account_age INT,
    total_call_minutes INT,
    total_sms_count INT,
    data_usage_gb FLOAT,
    monthly_charge FLOAT,
    customer_satisfaction_score INT,
    contract_type VARCHAR(10),
    number_of_complaints INT,
    is_premium_customer TINYINT,
    avg_call_duration FLOAT,
    number_of_recharges INT,
    international_calls_minutes INT,
    roaming_minutes INT,
    total_mms_count INT,
    total_data_plans INT,
    current_plan VARCHAR(10),
    number_of_referrals INT,
    churn_risk_score FLOAT,
    customer_support_calls INT,
    on_time_payments INT
);
"""
cursor.execute(create_table_query)

# Insert data into MySQL
insert_query = """
INSERT INTO telecom_data (
    customer_id, account_age, total_call_minutes, total_sms_count,
    data_usage_gb, monthly_charge, customer_satisfaction_score,
    contract_type, number_of_complaints, is_premium_customer,
    avg_call_duration, number_of_recharges, international_calls_minutes,
    roaming_minutes, total_mms_count, total_data_plans, current_plan,
    number_of_referrals, churn_risk_score, customer_support_calls,
    on_time_payments
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
for row in data:
    cursor.execute(insert_query, row)
    conn.commit()

# Step 3: Transform Data in MySQL
transform_query = """
SELECT
    current_plan,
    AVG(monthly_charge) AS avg_monthly_charge,
    SUM(total_call_minutes) AS total_call_minutes,
    SUM(data_usage_gb) AS total_data_usage,
    AVG(churn_risk_score) AS avg_churn_risk
FROM telecom_data
GROUP BY current_plan;
"""
cursor.execute(transform_query)

# Fetch transformed data
transformed_data = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]

# Convert Decimal objects to floats
transformed_data = [
    [float(value) if isinstance(value, Decimal) else value for value in row]
    for row in transformed_data
]

# Step 4: Write Transformed Data to JSON and CSV
json_file_path = 'transformed_telecom_data.json'  # Update this path as needed
csv_output_path = 'transformed_telecom_data.csv'  # Update this path as needed

# Write to JSON
with open(json_file_path, mode='w') as json_file:
    json.dump([dict(zip(columns, row)) for row in transformed_data], json_file, indent=4)

# Write to CSV
with open(csv_output_path, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(columns)  # Writing the header
    writer.writerows(transformed_data)  # Writing the data rows

# Close the cursor and connection
cursor.close()
conn.close()

print(f"Transformed data has been written to {json_file_path} and {csv_output_path}")
```
